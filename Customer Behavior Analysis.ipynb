```python
import pandas as pd
```

```python
orders = pd.read_csv("olist_orders_dataset.csv")
customers = pd.read_csv("olist_customers_dataset.csv")
order_items = pd.read_csv("olist_order_items_dataset.csv")
```

```python
df = orders.merge(customers, on='customer_id')
df.head()
```

```python
df = df.merge(order_items, on='order_id')
df.head()
```

```python
df['total_price'] = df['price'] + df['freight_value']
```

```python
customer_spending = df.groupby('customer_id')['total_price'].sum()
customer_spending.head()
```

```python
customer_spending.sort_values(ascending=False).head(10)
```

```python
customer_spending = customer_spending.reset_index()

customer_data = customer_spending.merge(customers, on='customer_id')

customer_data.head()
```

```python
customer_data.groupby('customer_city')['total_price'].sum().sort_values(ascending=False).head(10)
```

```python
orders_per_city = customer_data.groupby('customer_city')['customer_id'].count().sort_values(ascending=False).head(10)
orders_per_city
```

```python
avg_spending_city = customer_data.groupby('customer_city')['total_price'].mean().sort_values(ascending=False).head(10)
avg_spending_city
```

```python
df.isnull().sum()
```

```python
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
df['order_approved_at'] = pd.to_datetime(df['order_approved_at'])
df['order_delivered_carrier_date'] = pd.to_datetime(df['order_delivered_carrier_date'])
df['order_delivered_customer_date'] = pd.to_datetime(df['order_delivered_customer_date'])
```

```python
df.to_csv("final_data.csv", index=False)
```
